```python
# settlement 
# packages 

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb


# Import dataset
sales = pd.read_csv('sales.csv', parse_dates=["Order Date"], index_col="Order Date")
sales.columns = sales.columns.str.replace(' ','_')
```

```python
# Key point 2 : profit evolution per category 

# Profit per category
profit_cat = sales.groupby('Category').Profit.mean().reset_index()
profit_cat.rename(columns = {'Profit' : 'Mean_profit'}, inplace = True)
profit_cat
```

```python
profit_cat_year = sales.groupby(['Category', 'Order_Year']).Profit.mean().reset_index()
profit_cat_year
```

```python
profit_cat_year_pivot = profit_cat_year.pivot(
columns = 'Category',
index = 'Order_Year',
values = 'Profit')
profit_cat_year_pivot
```

```python
fig, ax = plt.subplots()

ax.plot(profit_cat_year_pivot.index, profit_cat_year_pivot['Furniture'])
ax.plot(profit_cat_year_pivot.index, profit_cat_year_pivot['Office Supplies'])
ax.plot(profit_cat_year_pivot.index, profit_cat_year_pivot['Technology'])
ax.set_xlabel("Time (years)")
ax.set_ylabel("Profit")
ax.set_title("Evolution of category profits")

plt.show()
```

```python
fig, ax = plt.subplots(3,1, sharey = False)

ax[0].plot(profit_cat_year_pivot.index, profit_cat_year_pivot['Furniture'], color = 'b', marker = 'v')
ax[1].plot(profit_cat_year_pivot.index, profit_cat_year_pivot['Office Supplies'], color = 'r', marker = 'o')
ax[2].plot(profit_cat_year_pivot.index, profit_cat_year_pivot['Technology'], color = 'g', marker = 'v', linestyle = '--')

ax[0].set_ylabel('Profit Fur')
ax[1].set_ylabel('Profit Off Supp')
ax[2].set_ylabel('Profit Tech')
ax[2].set_xlabel('Time (years)')

plt.show()
```

```python
# Profit for tables per segment

profit_table_segment = sales.groupby(['Sub-Category', 'Segment']).Profit.mean()
profit_table_segment = profit_table_segment.xs(key='Tables').reset_index()
profit_table_segment
```

```python
# Add values on chart
def values_on_bars(chart):
    for p in chart.patches:
        chart.annotate(format(p.get_height(), '.2f'), 
                       (p.get_x() + p.get_width() / 2., 
                       p.get_height()), 
                       ha = 'center', 
                       va = 'center',
                       xytext = (0, 10), 
                       textcoords = 'offset points')
```

```python
plt.figure( figsize = (10, 7) )
chart4=sb.barplot(data=profit_table_segment, x='Segment',y='Profit', ci=None)
chart4.set_title('Profit generated by the tables in each segment')
chart4.set_xlabel('Segment')
chart4.set_ylabel('Profit')
values_on_bars(chart4)
plt.show()
```

Every segment has a low profit for tables.

```python
# Profit for tables per ship mode
# Profit for tables
profit_table_ship_mode = sales.groupby(['Sub-Category', 'Ship_Mode']).Profit.mean()
profit_table_ship_mode = profit_table_ship_mode.xs(key='Tables').reset_index()
profit_table_ship_mode
```
